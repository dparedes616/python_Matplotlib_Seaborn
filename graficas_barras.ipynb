```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country=["India", "Japon", "Mexico", "Colombia", "Alemania"]
BIP=[100, 150, 50, 45, 160]
```

```python
plt.plot(country,BIP)
plt.bar(country,BIP, width=0.5, color=["red", "blue"])
plt.xticks(np.arange(5), ("IND", "JAP", "MEX", "COL", "GER"))
plt.show
```

```python
plt.barh(country,BIP)
plt.show
```
