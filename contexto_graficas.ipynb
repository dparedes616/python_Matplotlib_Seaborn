```python
import matplotlib.pylab as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=np.sin(x)
```

```python
fig, axes = plt.subplots(1,2, figsize=(5,5))
axes[0].plot(x,y, "r", label="Seno de X")
axes[0].set_title("Relación X y Y")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].legend()
axes[1].plot(y,x, "b", label="$sin(y)$")
axes[1].set_title("Relación Y y X")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Y")
axes[1].legend()
```

```python
plt.plot(x,y, label="sin(x)")
plt.title("Seno de X")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="lower left")
```
