```python
import matplotlib.pyplot as plt
import numpy as np

#Documentaci√≥n https://matplotlib.org/stable/plot_types/basic/plot.html#sphx-glr-plot-types-basic-plot-py
```

```python
x=np.linspace(0,5,11) #Crea array
print(x)
y=x**2
```

```python
plt.plot(x,y) 
plt.show()
plt.plot(x,y,"ro") #Primera letra color y el segundo es el marcador
plt.show()
plt.plot(x,y,"bo-") #Primera letra color, el segundo es el marcador y el tercera la continuidad
plt.show()
plt.plot(x,y,"bD:") #Primera letra color, el segundo es el marcador y el tercera la continuidad
plt.show()
plt.hist(x) #histogramas
plt.show()
plt.pie(x) #torta
plt.show()
plt.boxplot(x) #Grafica de bigotes
plt.show()
```
