```python
#Multi graficos
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
print(x)
y=x**2
print(y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.hist(y)
plt.show
#el primer valor de subplot es el numerode columnas que quiero, luego las filas y despues el indice, ubicacionde cada grafico
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.hist(y)
plt.show
```
