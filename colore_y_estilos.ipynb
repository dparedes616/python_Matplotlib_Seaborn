```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use("dark_background")
```

```python
fig, axes=plt.subplots(figsize=(8,8))
axes.plot(x,x+1,"r--")
axes.plot(x,x+2,"b*")
axes.plot(x,x+3,"y-*")
axes.plot(x,x+4,"w")
```

```python
fig, axes=plt.subplots(figsize=(8,8))
axes.plot(x,x+1,color="red", alpha=0.1, linewidth=5, linestyle="-", marker="o")
axes.plot(x,x+2,color="blue", alpha=0.8, linewidth=5, linestyle="--" )
axes.plot(x,x+3,color="#10BB00", alpha=0.5, linewidth=4, linestyle="dashed")
axes.plot(x,x+4,color="#CCFFCC")
##https://matplotlib.org/3.5.1/tutorials/colors/colors.html
```
